```python
import os
from dotenv import load_dotenv
from pathlib import Path

PROJECT_ROOT = Path(r"C:\Users\dunca\OneDrive\Documents\UON\SCS6104\Capstone")
load_dotenv(PROJECT_ROOT / ".env")

# Check MaxMind .mmdb path
mmdb_path_str = os.getenv("MAXMIND_DB_PATH")
print(f"MAXMIND_DB_PATH: {mmdb_path_str}")

if mmdb_path_str:
    mmdb_path = Path(mmdb_path_str)
    # If relative, resolve from project root
    if not mmdb_path.is_absolute():
        mmdb_path = PROJECT_ROOT / mmdb_path
    print(f"  Resolved: {mmdb_path}")
    print(f"  Exists:   {mmdb_path.exists()}")
    if mmdb_path.exists():
        print(f"  Size:     {mmdb_path.stat().st_size / 1e6:.1f} MB")

# Check AbuseIPDB key (don't print the key itself!)
api_key = os.getenv("ABUSEIPDB_API_KEY")
print(f"\nABUSEIPDB_API_KEY present: {bool(api_key)}")
print(f"  Length: {len(api_key) if api_key else 0}  (expected ~80 chars)")
```
